```python
## Problem Statement
##We want to search for a rating inside a list of Netflix ratings.
##We will use two algorithms:
##1. Linear Search
##2. Binary Search
```

```python
### Linear Search
##Linear search checks each element one by one until it finds the target.
```

```python
# Linear search checks each element one by one
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:   # compare each element with target
            return i           # return index if found
    return -1                  # return -1 if not found

# Example usage
ratings = [3, 5, 2, 4, 1]
print("Target found at index:", linear_search(ratings, 4))  # Output: 3
```

```python
### Binary Search
## Binary search is faster but requires the list to be sorted.
```

```python
#Binary search requires sorted data
def binary_search(arr, target):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low+high)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            low = mid +1 
        else:
            high = mid -1
    return
   # Example usage
ratings = sorted([3, 5, 2, 4, 1])  # must sort first
print("Target found at index:", binary_search(ratings, 4))  # Output: 3     
    
```
